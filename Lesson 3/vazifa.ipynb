```python
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
# 1. urta arifmetik usulni ishlatib 'longitude'ni tuldirish

df['longitude'].fillna(df['longitude'].mean(), inplace=True)
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
# 2. 'city'ni eng kup qaytarilgan qiymat bilan tuldirish

df['city'].fillna(df['city'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
# 3. 'full_address'ni eng kup kelgan qiymat bilan tuldirish

df['full_address'].fillna(df['full_address'].mode()[0], inplace=True) 
```

```python
df.isnull().sum()
```

```python
# 4. 'name'ni uzimiz ruchnoy tuldiramiz

df['name'] = df['name'].fillna('Ali')
```

```python
df.isnull().sum()
```

```python
df['name'].head(5)
```

```python
# 5. 'review_count'ni ustundi urtasidagi qiymat bilan tuldirish

df['review_count'] = df['review_count'].fillna(df['review_count'].median())
```

```python
df.isnull().sum()
```

```python
df['review_count'].head(20)
```

```python
# 6. NaN bor qatorlarni uchirish

df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
# 7. dataset da state juda kop tushub qolgani uchun bu ustundi uchirvoramiz

df.drop('state', axis=1, inplace=True)
```

```python
df.head(30)
```

```python
# 8. barcha 'type' lar business bulgani uchun, tushurib qoldirilgan datalarniham 'business' qilib quyamiz

df['type'] = df['type'].fillna('business')
```

```python
# Task 1: Check Voting Eligibility, Ask the user for their age. If they are 18 or older, print "You are eligible to vote." Otherwise, print "You are not eligible to vote."

x = int(input("how old are you: "))

if x >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# Task 2: Find the Biggest Number, Ask the user for three numbers. Print out which number is the biggest.

x1 = int(input("write a first number"))
x2 = int(input("write a second number"))
x3 = int(input("write a third number"))

biggest_number = x1

if x2 > biggest_number:
    biggest_number = x2
if x3 > biggest_number:
    biggest_number = x3

print(f"most biggest number is: {biggest_number}")
```

```python
# Task 3: Password Strength Check, Ask the user to input a password. If the password length is at least 8 characters, print "Strong password". Otherwise, print "Weak password".

x = str(input("write your password: "))

if len(x) >= 8:
    print("strong password!")
else:
    print("weak password")
```

```python
''' Task 4: Grade Evaluation 
Ask the user for their exam score (0-100).  
- Print "Excellent" if score is 90 or above.
- Print "Good" if score is between 70 and 89.
- Print "Needs Improvement" if score is below 70. '''

x = int(input("write your score: "))

if x >= 90:
    print("Excellent")
elif x >= 70 and x <= 89:
    print("Good")
else:
    print("Needs Improvement")
```

```python
''' Task 5: Divisible by 3 and 5
Ask the user for a number.  
- If it's divisible by both 3 and 5, print "FizzBuzz".
- If only divisible by 3, print "Fizz".
- If only divisible by 5, print "Buzz".
- Otherwise, print the number. '''

x = int(input("write a number: "))

if x % 3 == 0 and x % 5 == 0:
    print("FizzBuzz")
elif x % 3 == 0:
    print("Fizz")
elif x % 5 == 0:
    print("Buzz")
else:
    print(x)
```
